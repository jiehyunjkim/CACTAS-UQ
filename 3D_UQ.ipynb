```python
npz_dir = "/raid/mpsych/CACTAS/DATA/nnUNet/Mask/UQ_v1"        # 3D
gt_dir = "/raid/mpsych/CACTAS/DATA/nnUNet/Mask/nnUNet_raw/Dataset001_ESUS/labelsTs"
img_dir = "/raid/mpsych/CACTAS/DATA/nnUNet/Mask/nnUNet_raw/Dataset001_ESUS/imagesTs"
```

```python
# 1. Compute baseline metrics: IoU and Expected Calibration Error (ECE)
# 2. Apply uncertainty-based rejection and recompute IoU and ECE
# 3. Report extended performance metrics before and after rejection 
#    (IoU, Dice, F1-score, and confusion matrix components)
# 4. Visualize representative plaque cases:
#    - If plaque exists unilaterally, display one example
#    - If bilateral plaques are present, visualize both sides
# 5. Illustrate voxel-level FP/FN distributions and their changes after rejection
```

```python
# 3D - Uncertainty Quantification and Visualization Pipeline
```

```python
npz_list = sorted(glob.glob(os.path.join(npz_dir, "*.npz")))
all_base_iou, all_reject_iou, all_ece_before, all_ece_after = [], [], [], []

for npz_path in npz_list:
    case = os.path.basename(npz_path).replace(".npz", "")
    gt_path = os.path.join(gt_dir, f"{case}.nrrd")
    cta_path = os.path.join(img_dir, f"{case}_0000.nrrd")
    
    case_num = os.path.basename(npz_path).split('.')[0]
    print(f"\n========== Case {case_num} ===========")
    
    probs = C.Uncertain.load_data(npz_path)
    base_iou, gt_bin, probs_gt, entropy_map = C.Uncertain.entropy_map(npz_path, gt_path, probs)
    flat_probs, flat_entropy, flat_labels, flat_preds = C.Uncertain.uncertainty_calc(probs_gt, entropy_map, gt_bin)
    base_ece_all, base_ece_fg = C.Uncertain.calc_ece(probs_gt, gt_bin, base_iou)
    keep, iou_reject, ece_after_fg = C.Uncertain.reject(flat_entropy, flat_preds, flat_labels, flat_probs)
    before, after = C.Uncertain.print_metrix(keep, flat_preds, flat_labels)
    C.Uncertain.visualize_uncertainty_map(probs_gt,entropy_map,cta_path,gt_bin, case_num)
    C.Uncertain.visualize_rejection(before, after, case_num)
    
    
    all_base_iou.append(base_iou)
    all_reject_iou.append(iou_reject)
    all_ece_before.append(base_ece_fg)
    all_ece_after.append(ece_after_fg)
```

```python
C.Uncertain.summary(all_base_iou, all_reject_iou, all_ece_before, all_ece_after)
```
